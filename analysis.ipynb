```python
import pandas as pd
import numpy as np
from ggplot import *
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
graduates = pd.read_csv("grad-students.csv")

recent_grads2 = recent_grads.set_index("Major")
graduates2 = graduates.set_index("Major")



unemployed_grad = graduates2['Grad_unemployment_rate']
unemployed = recent_grads2['Unemployment_rate']

unemployed = unemployed.to_frame()
unemployed_grad = unemployed_grad.to_frame()



#unemployed.plot(kind='bar')
unemployed_grad.plot(kind='bar')
```

```python
engineering = recent_grads[recent_grads['Major_category'] == 'Engineering']
grad_engineering = graduates[graduates['Major_category'] == 'Engineering']

altered_engineering = engineering.set_index("Major")
#altered_engineering.plot(y="Unemployment_rate", kind="bar")

altered_gradengineering = grad_engineering.set_index("Major")
#altered_gradengineering.plot(y="Grad_unemployment_rate", kind="bar")


print(engineering.Major.unique())

diff = engineering.Unemployment_rate.mean() - grad_engineering.Grad_unemployment_rate.mean()
print(str(diff))

for i in engineering.Major.unique():
    unem_diff = altered_engineering.Unemployment_rate[i] - altered_gradengineering.Grad_unemployment_rate[i]
    print("Unemployment difference for " + i + ": " + str(unem_diff))
```

```python
table1 = recent_grads.groupby(['Major_category'])
table2 = graduates.groupby(['Major_category'])

table1_unemployment = table1.Unemployment_rate.agg(np.mean)
table2_unemployment = table2.Grad_unemployment_rate.agg(np.mean)

table1_salary = table1.Median.agg(np.mean)
table2_salary = table2.Grad_median.agg(np.mean)


table3_unemp = table1_unemployment-table2_unemployment
table3_salary = table2_salary - table1_salary

#table1.plot(y="Unemployment_rate", kind="bar")
table3_unemp.plot(kind="bar")
```

```python
stem_majors = ['Engineering', 'Physical Sciences', 'Biology & Life Science', 'Computers & Mathematics', 'Psychology & Social Work']

frames = []
for i in stem_majors:
    frames.append(recent_grads[recent_grads['Major_category'] == i])

result = pd.concat(frames)
result
```

```python
stem_med = result.Median
stem_med.describe()
```

```python
all_majors = recent_grads.Major_category.unique()
new_frames = []
for i in all_majors:
    if i not in stem_majors:
        new_frames.append(recent_grads[recent_grads['Major_category'] == i])

non_stem_result = pd.concat(new_frames)
non_stem_result
```

```python
non_stem_med = non_stem_result.Median
non_stem_med.describe()
```
