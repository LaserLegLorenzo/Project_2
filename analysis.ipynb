```python
import pandas as pd
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
graduates = pd.read_csv("grad-students.csv")
recent_grads2 = recent_grads.set_index("Major")
graduates2 = graduates.set_index("Major")

unemployed_grad = graduates2['Grad_unemployment_rate'].head()
unemployed = recent_grads2['Unemployment_rate'].head()

unemployed = unemployed.to_frame()
unemployed_grad = unemployed_grad.to_frame()


unemployed.plot(kind='bar')
unemployed_grad.plot(kind='bar')
```
